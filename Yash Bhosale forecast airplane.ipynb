```python
train = data.head(80)
test = data.tail(16)
train = train.astype('float32')
test = test.astype('float32')
```

```python
import itertools
```

```python
p = range(0,5)
d = range(0,3)
q = range(0,5)
params = list(itertools.product(p,d,q))
```

```python
resid = []
param = []
for pdq in params:
    try:
        model = ARIMA(train,order = pdq).fit()
        preds = model.predict(start = test.index[0],end = test.index[-1])
        error = np.sqrt(mean_squared_error(preds,test))
        resid.append(error)
        param.append(pdq)
        
    except:
        continue
```

```python
pdq = pd.DataFrame(resid,param)
```

```python
# Simple Exponential Model
```

```python
ses = SimpleExpSmoothing(train).fit()
ses_preds = ses.predict(start = test.index[0],end = test.index[-1])
ses_rms = np.sqrt(mean_squared_error(ses_preds,test))
ses_rms
```

```python
ses_preds.plot()
test.plot()
```

```python
# Holt's Model (Double Exoponential) Trend
```

```python
holt = Holt(train).fit()
holt_preds = holt.predict(start = test.index[0],end = test.index[-1])
holt_rms = np.sqrt(mean_squared_error(holt_preds,test))
holt_rms
```

```python
holt_preds.plot()
test.plot()
```

```python
# Winters Method (Advanced Exponential Method) Additive Seasonality + Additive Trend 
```

```python
Winter_aa = ExponentialSmoothing(train,seasonal= 'add',trend = 'add',seasonal_periods=12).fit()
Winter_aa_preds = Winter_aa.predict(start = test.index[0],end = test.index[-1])
Winter_aa_rms = np.sqrt(mean_squared_error(Winter_aa_preds,test))
Winter_aa_rms
```

```python
Winter_aa_preds.plot()
test.plot()
```
